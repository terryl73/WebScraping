```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://mars.nasa.gov/news/'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
#Find Title

result = soup.find_all('div', class_='content_title')

Block1 = result[0]

title = Block1.find('a').text

print(title)

#for title in title_results:
 #   print(title)
```

```python
#Find Paragraph Text

link = Block1.a['href']
url = "https://mars.nasa.gov" + link
print(url)

response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```

```python
p_result = soup.find_all('p')
for i in range(0,5):
    print(p_result[i])
    print("----------------")
    
news_p = p_result[1].text
print(news_p)
```

```python
from splinter import Browser
from splinter.exceptions import ElementDoesNotExist
from bs4 import BeautifulSoup
```

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
browser.visit(url)

html = browser.html
soup = BeautifulSoup(html, 'html.parser')

print(soup.prettify())
```

```python
#Find NASA image link
base_url = "https://www.jpl.nasa.gov/"
pic_soup = soup.find('a', class_="button fancybox")
print(pic_soup)

pic_link = pic_soup['data-fancybox-href']
pic_link

featured_image_url = base_url + pic_link

browser.visit(featured_image_url)

#="/spaceimages/details.php?id=PIA17843
```

```python
#Mars Weather

url = 'https://twitter.com/marswxreport?lang=en'
response = requests.get(url)
soup = BeautifulSoup(response.text, 'html.parser')
print(soup.prettify())
```
